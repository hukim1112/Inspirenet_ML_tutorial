```python
boston = load_boston()
print(boston.data.shape)
```

```python
data = boston.data
name = boston.feature_names
target = boston.target
```

```python
x_0 = np.ones(len(target))
col_2 = data[:,2]
col_4 = data[:,4]
col_5 = data[:,5]
col_6 = data[:,6]
col_12 = data[:,12]
```

```python
col_list = [col_2, col_4, col_5, col_6, col_12]
```

```python
def mean_normalization(x):
    return (x-x.mean()) / (x.max()-x.min())
```

```python
scaled_col = list(map(lambda x: mean_normalization(x), col_list))
```

```python
param = np.random.normal(size=6)
```

```python
scaled_col.insert(0, x_0)
```

```python
features = np.array(scaled_col)
```

```python
learning_rate = 0.01
iteration = 4000
num_example = len(target)
```

```python
def hypothesis(x, param):
    pred = np.dot(np.transpose(x), param)
    return pred
```

```python
def cost_func(x, param):
    diff = hypothesis(x, param) - target
    return 1/(2*num_example)*np.dot(diff, diff)
```

```python
def derivative(x, param):
    return (1/num_example)*np.dot(x, hypothesis(x, param) - target) 
```

```python
def update(x, param):
    return param-learning_rate*derivative(x,param)
```

```python
for i in range(iteration):
    print (cost_func(features, param))
    param = update(features, param)
    
```
